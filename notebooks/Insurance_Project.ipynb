โจทย์ธุรกิจ:

บริษัทประกันอยากรู้ว่าลูกค้ากลุ่มไหนควรดูแลก่อน เพราะอาจมี retention risk จากเบี้ยประกันที่หนัก รายได้ไม่สูง หรือ coverage exposure สูง

```python
#เพิ่มจากที่หน่วยเดือนให้กลายเป็นหน่วย ปี
df["annual_premium"] = df["monthly_premium"] * 12
#ดูว่าจากรายได้นั้นเทียบกับเบี้ยประกันเป็นอย่างไรบ้าง เบี้ยกินกี่ % ของรายได้
df["annual_premium_to_income_ratio"] = df["annual_premium"] / df["annual_income"]
#ประกันจ่ายเยอะไหมเทียบกับรายได้ เอาไว้ดูเพื่อที่จะดูว่าคนกลุ่มนี้มีรายได้เยอะแค่ไหน เทียบกับที่เค้าจะได้วงเงินที่บริษัทคุ้มครอง
df["coverage_to_income_ratio"] = df["coverage_amount"] / df["annual_income"]
#ประกันจ่ายเทียบกับเบี้ยประกันที่ลูกค้าจ่าย เอาไว้ดูเพื่อที่จะประเมินว่าคนกลุ่มนี้จ่ายเงินเยอะไหมเทียบกับที่บริษัทจ่าย
df["coverage_to_annual_premium_ratio"] = df["coverage_amount"] / df["annual_premium"]
```

```python
#ยิ่ง premium_to_income น้อยก็จะยิ่งดี
#ยิ่ง coverage_to_income น้อยก็จะยิ่งดี
#ยิ่ง coverage_to_premium น้อยก็จะยิ่งดี
df[[
    "annual_income",
    "monthly_premium",
    "annual_premium",
    "premium_to_income_ratio",
    "coverage_amount",
    "coverage_to_income_ratio",
    "coverage_to_premium_ratio"
]].head()
```

premium_to_income : max สุดไม่เกิน 20% ของรายได้ จุดตัดน่าจะอยู่ที่ 5% ของรายได้
coverage_to_income : ส่วนมากแล้วจะได้รับการคุ้มครองอยู่ที่ประมาณ 4.5 เท่า ของรายได้ แต่ max อยู่ที่ 28+ เท่า
coverage_to_premium : ขั้นต่ำอยู่ที่ 12 เท่า แต่มากที่สุดอยู่ที่ ประมาณ 4000 เท่า และ mean อยู่ที่ 256 เท่า

```python
#distribution-based cutoff

q1 = df["premium_to_income_ratio"].quantile(0.25)
q3 = df["premium_to_income_ratio"].quantile(0.75)
iqr = q3 - q1

elevated_cutoff = q3
high_cutoff = q3 + 1.5 * iqr

print(elevated_cutoff, high_cutoff)


def label_affordability_risk(ratio):
    if ratio <= elevated_cutoff:
        return "Standard burden"
    elif ratio <= high_cutoff:
        return "Elevated burden"
    else:
        return "High burden watchlist"

df["affordability_risk"] = df["premium_to_income_ratio"].apply(label_affordability_risk)

```

ดังนั้น cutoff ที่ defend ได้ดีกว่าคือ:

Segment	Rule	เหตุผล
Standard burden	<= 2.65%	อยู่ใน 75% แรกของ portfolio
Elevated burden	> 2.65% ถึง <= 5%	สูงกว่าลูกค้าส่วนใหญ่ แต่ยังไม่ใช่ tail หนัก
High burden watchlist	> 5%	เกิน robust upper threshold ใกล้ outlier zone

```python
coverage_q3 = df["coverage_to_income_ratio"].quantile(0.75)
coverage_p90 = df["coverage_to_income_ratio"].quantile(0.90)

def label_coverage_exposure(ratio):
    if ratio <= coverage_q3:
        return "Standard exposure"
    elif ratio <= coverage_p90:
        return "Elevated exposure"
    else:
        return "High exposure watchlist"

df["coverage_exposure"] = df["coverage_to_income_ratio"].apply(label_coverage_exposure)
```

```python
# บริษัทรับ exposure สูงแค่ไหน
df["coverage_exposure"].value_counts()
```

ลูกค้าส่วนใหญ่ หรือประมาณ 75% มี coverage อยู่ในระดับปกติของ portfolio
ประมาณ 15.6% มี coverage สูงกว่าลูกค้าส่วนใหญ่
ประมาณ 9.3% เป็นกลุ่ม coverage สูงมากเมื่อเทียบกับรายได้

บริษัทมีทุนประกันที่ต้องรับผิดชอบสูงมากเมื่อเทียบกับรายได้ของลูกค้ากลุ่ม High exposure watchlist

```python
def label_portfolio_priority(row):
    if (
        row["affordability_risk"] == "High burden watchlist"
        and row["coverage_exposure"] == "High exposure watchlist"
    ):
        return "Priority 1 - high burden and high exposure"

    elif row["affordability_risk"] == "High burden watchlist":
        return "Priority 2 - high affordability burden"

    elif row["coverage_exposure"] == "High exposure watchlist":
        return "Priority 3 - high coverage exposure"

    elif row["affordability_risk"] == "Elevated burden":
        return "Priority 4 - elevated affordability burden"

    elif row["coverage_exposure"] == "Elevated exposure":
        return "Priority 5 - elevated coverage exposure"

    else:
        return "Routine monitoring"

df["portfolio_priority"] = df.apply(label_portfolio_priority, axis=1)
```

## Key Findings

1. A focused high-priority segment was identified.
   The analysis identified 348 Priority 1 policyholders with both high premium burden and high coverage exposure. This is a small but important group for proactive retention monitoring.

2. Premium burden is concentrated in the upper tail.
   Premium-to-income ratio is right-skewed. Most policyholders have low premium burden, while the high-burden watchlist sits above the robust upper threshold based on Q3 + 1.5 IQR.

3. Priority 1 is concentrated in specific products and income segments.
   Priority 1 policyholders are concentrated in Universal Life and Whole Life products, especially among customers earning below 80k annually.

4. Customer profile adds supporting context.
   Priority 1 customers tend to be older, have more dependents, and are more likely to be married than Routine Monitoring customers. However, the main priority logic is driven by affordability and exposure, not demographics alone.


## Business Recommendations


1. Prioritize Priority 1 policyholders for proactive retention review.
   These customers have both high premium burden and high coverage exposure, making them the most important group for early monitoring.

2. Review Universal Life and Whole Life policies first.
   Priority 1 customers are concentrated in these two policy types, so retention actions should be product-specific rather than applied equally across all products.

3. Focus affordability support on lower-income customers.
   Most Priority 1 customers earn below 80k annually, suggesting that premium burden is a key driver of retention concern.

4. Use premium-to-income and coverage-to-income ratios as monitoring KPIs.
   These ratios allow the business to track affordability pressure and exposure risk even without an actual lapse target variable.
